```python
import gensim
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
big_bang_thory_df=pd.read_csv(r"bigbangtheory.csv")
```

```python
nltk.download('punkt')
```

```python
simple_preprocess(str(sent_tokenize(big_bang_thory_df['dialogue'][0])))
```

```python
big_bang_thory_df.shape[0]
```

```python
big_bang_thory_df.reset_index(inplace=True)
```

```python
total_dialogue=[]
for i in range(len(big_bang_thory_df['dialogue'])):
    for j in sent_tokenize(big_bang_thory_df['dialogue'][i]):
        total_dialogue.append(simple_preprocess(j))
        
    
```

```python
total_dialogue
```

```python
model=Word2Vec(window=5,min_count=2)
```

```python
model.build_vocab(total_dialogue)
```

```python
model.wv.index_to_key ### to see the vocabulary
```

```python
## train the model
model.train(total_dialogue,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv['leonard'] ### see here it is case sensitive
```

```python
model.wv.most_similar('leonard')
```

```python
model.wv.doesnt_match(['leonard','penny','raj','priya']) ### the lowest similarity
```

```python
model.wv.get_normed_vectors()
```
